```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd

tp = pd.read_csv('prev_tp.csv')
stop = pd.read_csv('prev_stop.csv')

tp_values   = tp.iloc[:, 1]
stop_values = stop.iloc[:, 1]
bins = np.arange(0, 0.16 + 0.001, 0.001)

# Histogramı çiz
plt.figure(figsize=(25, 12))
plt.hist([tp_values, stop_values], bins=bins, label=['tp', 'stop'])
plt.legend()
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.title('tp vs stop Histogram')
plt.show()
```

```python
import numpy as np
import pandas as pd


counts_tp,   bin_edges = np.histogram(tp_values,   bins=bins)
counts_stop, _         = np.histogram(stop_values, bins=bins)

print("Bin width  |  tp frequency  |  stop frequency")
print("------------------------------------------")
for i in range(len(counts_tp)):
    left  = bin_edges[i]
    right = bin_edges[i+1]
    print(f"{left:.3f}–{right:.3f} |      {counts_tp[i]:3d}      |      {counts_stop[i]:3d}")
```
