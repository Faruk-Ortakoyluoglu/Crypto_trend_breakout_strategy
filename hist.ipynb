```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd

tp = pd.read_csv('tp.csv')
stop = pd.read_csv('stop.csv')

# İlk sütunu al (tek kolonluysa direkt o seriyi kullan)
tp_values   = tp.iloc[:, 1]
stop_values = stop.iloc[:, 1]

custom_bins = np.arange(0, 0.16 + 0.001, 0.001)

# Histogramı çiz
plt.figure(figsize=(25, 12))
plt.hist([tp_values, stop_values], bins=custom_bins, label=['tp', 'stop'])
plt.legend(loc='upper right')
plt.xlabel('Değer')
plt.ylabel('Frekans')
plt.title('tp vs stop Histogramı')
plt.show()
```
